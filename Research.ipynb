<h3>Какую боль чувствуют люди при искривлении спины?</h3>

```python
import pandas as pd

# Загрузка данных из CSV файла в DataFrame
data = pd.read_csv("Новая форма.csv")

# Убедитесь, что столбец с информацией о симптомах называется правильно
# Здесь предполагается, что он называется '17. Какие начальные симптомы искривления спины вы можете почувствовать на себе?'
symptoms_column = data['17. Какие начальные симптомы искривления спины вы можете почувствовать на себе?  ']

# Разделение строк на отдельные симптомы
symptoms = symptoms_column.str.split(';')

# Создание пустого списка для хранения всех симптомов
all_symptoms = []

# Итерация по каждой строке и добавление симптомов в список
for symptom_list in symptoms:
    all_symptoms.extend(symptom_list)

# Преобразование списка в Series для удобства подсчета частоты
all_symptoms_series = pd.Series(all_symptoms)

# Подсчет частоты каждого симптома
symptom_counts = all_symptoms_series.value_counts()

# Вывод частоты симптомов
print("Частота разных симптомов искривления спины:")
print(symptom_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Новая форма.csv")


symptoms_column = data['17. Какие начальные симптомы искривления спины вы можете почувствовать на себе?  ']


symptoms = symptoms_column.str.split(';')
all_symptoms = []

for symptom_list in symptoms:
    all_symptoms.extend(symptom_list)

all_symptoms_series = pd.Series(all_symptoms)


symptom_counts = all_symptoms_series.value_counts()


plt.figure(figsize=(10, 6))
symptom_counts.plot(kind='bar')
plt.title("Частота разных симптомов искривления спины")
plt.xlabel("Симптомы")
plt.ylabel("Частота")
plt.xticks(rotation=90)  

plt.show()
```

<h1>Теперь сравним кто из мужчин или женщин чаще занимается спортом</h1>

```python
import pandas as pd


data = pd.read_csv("Новая форма.csv")  



male_active = data[(data['2. Пол: '] == "Мужской") & (data["15. Как часто вы упражняетесь или занимаетесь физической активностью? "] == "1-2 раза в неделю")]
female_active = data[(data['2. Пол: '] == "Женский") & (data['15. Как часто вы упражняетесь или занимаетесь физической активностью? '] == "1-2 раза в неделю")]


percent_male_active = (len(male_active) / len(data[data['2. Пол: '] == "Мужской"])) * 100
percent_female_active = (len(female_active) / len(data[data['2. Пол: '] == "Женский"])) * 100

print("Процент мужчин, занимающихся физической активностью:", percent_male_active, "%")
print("Процент женщин, занимающихся физической активностью:", percent_female_active, "%")
```

<h3>Pie Chart</h3>

```python
import matplotlib.pyplot as plt


labels = ['Мужчины', 'Женщины']
sizes = [percent_male_active, percent_female_active]
colors = ['lightblue', 'lightcoral']
explode = (0.1, 0)  


plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal') 

plt.title('Процент занимающихся физической активностью среди мужчин и женщин в 1-2 раза в неделю ')
plt.show()
```
